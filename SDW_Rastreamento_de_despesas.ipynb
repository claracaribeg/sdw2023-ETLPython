## **Passos do Projeto:**


### **Extract (E)**

Extraia o arquivo criado. Este código lê o arquivo CSV e carrega os dados em uma estrutura tabular, tornando-os prontos para análise e manipulação adicionais. Certifique-se de substituir 'nome_do_arquivo.csv' pelo nome do seu arquivo CSV real.


````python
import pandas as pd

# Substitua 'despesas_set.csv' pelo nome real do seu arquivo CSV
despesas_set = 'despesas_set.csv'
dados_despesas = pd.read_csv(despesas_set)

# Examine os dados (opcional)
print(dados_despesas.head())
````

### **Transform (T)**

Neste passo, iremos calcular o total gasto em cada categoria de despesa. Para fazer isso, utilizaremos os dados do arquivo CSV que foram lidos no passo anterior. A ideia é agrupar os dados por categoria e, em seguida, somar os valores gastos em cada uma delas. Isso nos permitirá obter um resumo claro de quanto foi gasto em cada categoria, que será útil para a análise financeira.



````python
total_por_categoria = dados_despesas.groupby('Categoria')['Valor'].sum()
print(total_por_categoria)
````

### **Load (L)**

Nesta etapa, após a leitura e transformação dos dados do arquivo CSV, vamos realizar a carga dos resultados em uma visualização gráfica. Usaremos a biblioteca Matplotlib para criar um gráfico de barras que representará o total gasto em cada categoria de despesa. Essa visualização facilitará a compreensão dos padrões de gastos e ajudará na análise financeira. O gráfico será gerado e exibido na tela, permitindo que você o interprete e tome decisões informadas com base nos seus dados de despesas.

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
total_por_categoria.plot(kind='bar', color='darkred')

plt.title('Total Gasto por Categoria de Despesa')
plt.xlabel('Categoria')
plt.ylabel('Total Gasto (R$)')  # Adicione a unidade monetária R$
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

for i, valor in enumerate(total_por_categoria):
    plt.text(i, valor + 5, f'R${valor:.2f}', ha='center', va='bottom', fontsize=10, color='black')

plt.show()
````
